```python
import pandas as pd
import os
import requests
from tqdm import tqdm
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Paths
csv_file = "/content/drive/MyDrive/cleaned_object_title_img_id.csv"  # Path to your dataset in Google Drive
image_save_dir = "/content/drive/MyDrive/images_complete/"  # Directory to save downloaded images in Google Drive

# URL template
url_template = "https://collections.carlos.emory.edu/internal/media/dispatcher/{}"

# Create the directory if it doesn't exist
os.makedirs(image_save_dir, exist_ok=True)

# Load the dataset
data = pd.read_csv(csv_file)

# Download all images
for idx, row in tqdm(data.iterrows(), total=len(data)):
    image_id = row["Image ID"]
    title = row["Title"]

    # Generate the URL
    image_url = url_template.format(image_id)
    image_path = os.path.join(image_save_dir, f"{image_id}.jpg")

    try:
        # Fetch the image
        response = requests.get(image_url, stream=True)
        if response.status_code == 200:
            # Save the image
            with open(image_path, "wb") as f:
                f.write(response.content)
        else:
            print(f"Failed to download image with ID {image_id}: {response.status_code}")
    except Exception as e:
        print(f"Error downloading image with ID {image_id}: {e}")

print("All images have been downloaded to Google Drive.")
```
