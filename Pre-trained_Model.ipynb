```python
import torch

if torch.backends.mps.is_available():
    print("MPS is available! Using the MPS backend.")
else:
    print("MPS is not available. Using CPU.")
```

```python
from diffusers import StableDiffusionPipeline
import os

# Check for MPS availability
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")

# Load the pretrained model with float16 precision
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
)
pipe.to(device)

# Enable memory optimization
pipe.enable_attention_slicing()

# Reduce image resolution
pipe.scheduler.config.height = 256
pipe.scheduler.config.width = 256

# Generate an image with a fixed seed
prompt = "A beautiful ancient artifact"
generator = torch.manual_seed(42)  # Optional: for reproducibility
result = pipe(prompt, generator=generator)

# Get the image
image = result.images[0]

# Show and save the image
image.show()
```
